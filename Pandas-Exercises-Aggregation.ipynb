# Aggregations

### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset occupation.csv from the folder or [here](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
users = pd.read_csv('occupation.csv', sep='|')
```

### Step 3. Assign it to a variable called users.

```python
users.info()
```

```python
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(['occupation'])[['age']].mean().add_prefix('Avg_').reset_index()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
users['male_encode'] = np.where(users.gender == 'M', 1,0)
```

```python
Total_male_occupation = users.groupby('occupation')['male_encode'].sum()
```

```python
Total_count = users.occupation.value_counts()
```

```python
Male_ratio = (Total_male_occupation/Total_count)*100
```

```python
Male_ratio.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg(['min','max'])
```

```python
max_Age = users.groupby(['occupation'])[['age']].max().add_prefix('Max_').reset_index()
min_Age = users.groupby(['occupation'])[['age']].min().add_prefix('Min_').reset_index()
```

```python
combined = pd.merge(min_Age, max_Age)
combined
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender'])[['age']].mean().add_prefix('Avg_').reset_index()
```

### Step 8.  For each occupation present the percentage of women and men

```python
pd.crosstab(users.occupation, users.gender, normalize='index').mul(100).round(2)
```
